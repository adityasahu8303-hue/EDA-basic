```python
import pandas as pd
df = pd.read_csv("/content/House Price India.csv")
df
```

```python
# there is space between column name so we replace it by underscore.
list[df.columns]
```

```python
df.columns = df.columns.str.strip()              # extra spaces remove
df.columns = df.columns.str.lower()              # lowercase
df.columns = df.columns.str.replace(" ", "_")    # space -> underscore

df.columns
```

```python
#Ques1 . house price  avg, median and stamndard deviation..We use aggregate function when we need to apply multiple aggregation at a time...
df["price"].agg(["mean", "median", "std" ])
```

```python
df["number_of_bedrooms"].value_counts()
```

```python
# Ques 2 .. Most common bedrooms...
df["number_of_bedrooms"].mode()
```

```python
# ques 3. check the price data is skewed..

''' yes the price data is highly positively right skewed...'''
df["price"].skew()
```

```python
# ques 4. find average of each number of bedrooms.
df.groupby("number_of_bedrooms")["price"].mean().round(2)
```

```python
df[["living_area", "price"]].corr()
```

```python
df["living_area"].corr(df["price"])
```

```python
# ques 5 Q5.  What is the relationship between living area and price? (Use : Area without Basement).
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.scatter(df["living_area"], df["price"], alpha=0.5)
plt.xlabel("Living Area (Without Basement)")
plt.ylabel("House Price")
plt.title("Living Area vs House Price")
plt.show()
```

```python
# Ques 6. Q6. Identify any anomalies where houses have high prices but low areas.
high_price = df["price"].quantile(0.80)
low_area = df["living_area"].quantile(0.20)

anomalies = df[(df["price"]>= high_price) & (df["living_area"]<= low_area )]
anomalies
```

```python
# Q7. Compare average price based on number of floors and houses with or without waterfront.
df.groupby(["number_of_floors", "waterfront_present"])["price"].mean().round(2).unstack()
```

```python
#Q8. Identify the minimum and maximum house price. What does this indicate?
df["price"].agg(["min", "max"])
```

**The minimum house price is 75,000 and the maximum house price is 7,700,000. This indicates a very wide price range in the dataset, suggesting significant variation in house values. The high maximum price may represent luxury properties or premium locations, while the minimum price represents lower-priced houses.**

```python
#Q9. Which location (zipcode/area) has the highest average price?
df.groupby("postal_code")["price"].mean().sort_values(ascending = False).head(10).round(2)
```

**Insight 1**

The house price distribution is highly positively (right) skewed (skewness ≈ 4), indicating that a few luxury houses have very high prices while most houses are moderately priced.

Insight 3 **bold text**

**Insight 4**

Houses with a waterfront have a higher average price than houses without a waterfront, showing that waterfront access significantly increases property value.

Insight 5
